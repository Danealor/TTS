```python
from pathlib import Path
 
path_in = 'raw_data/'
path_out = 'dataset/wavs/'

files = Path(path_in).glob('*.ogg')
for file in files:
    print(file)
```

```python
import os
import soundfile as sf

os.makedirs(path_out, exist_ok=True)
files = Path(path_in).glob('*.ogg')
for file in files:
    data, samplerate = sf.read(file)
    outfile = Path(path_out).joinpath(file.name).with_suffix('.wav')
    sf.write(outfile, data, 22050)
    print(f'{file} ({samplerate} Hz) -> {outfile} (22050 Hz)')
```

```python
import pandas as pd

df = pd.read_excel('dataset/metadata.xlsx', header=None, names=['filename', 'text'])
df
```

```python
files_to_remove = list(df.loc[df['text'] == 'X', 'filename'])
files_to_remove
```

```python
import os

files = Path(path_out).glob('*.wav')
for file in files:
    if file.stem in files_to_remove:
        os.remove(file)
        print(f"Deleted {file}")
```

```python
import csv

filtered_df = df[df['text'] != 'X']
filtered_df.to_csv('dataset/metadata.txt', sep='|', header=False, index=False, quoting=csv.QUOTE_NONE)
```
