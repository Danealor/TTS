```python
from TTS.utils.synthesizer import Synthesizer

synthesizer = Synthesizer(
    model_path,
    config_path,
    speakers_file_path,
    language_ids_file_path,
    use_cuda=True,
)
```

```python
synthesizer.tts_model.speaker_manager.load_embeddings_from_file(speaker_embeddings)
list(synthesizer.tts_model.speaker_manager.name_to_id.keys())
```

```python
import matplotlib.pyplot as plt
import librosa.display
import IPython.display as ipd
import numpy as np

audio = np.array(synthesizer.tts("It's only fun when they run!", "VCTK_p286", "en"))

sr = synthesizer.tts_model.config.audio.sample_rate
plt.figure(figsize=(14, 5))
librosa.display.waveplot(audio, sr=sr)
ipd.Audio(audio, rate=sr)
```
